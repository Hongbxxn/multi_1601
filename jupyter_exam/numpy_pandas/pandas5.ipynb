```python
date_list = [
    {'yyyy-mm-dd':'2003-04-18'},
    {'yyyy-mm-dd':'2010-05-12'},
    {'yyyy-mm-dd':'2020-07-27'}    
]
df = pd.DataFrame(date_list, columns=['yyyy-mm-dd'])
df
```

```python
def get_year(column):
    return column.split('-')[0]
get_year('2002-09-09')
```

```python
df['year'] = df['yyyy-mm-dd'].apply(get_year)
df
```

```python
df['year_map'] = df['yyyy-mm-dd'].map(get_year)
```
