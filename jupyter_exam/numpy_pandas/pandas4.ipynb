### 숫자형 데이터 vs 범주형 데이터 

```python
import pandas as pd
df_train = pd.read_csv('train.csv')
df_train.info()
```

```python
df_train['Pclass'] = df_train['Pclass'].astype(str)
```

```python
df_train.info()
```

```python
df_train['Age']
```

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df_train['Age'].apply(age_categorize)
```

```python
### group by 

pclass_group = df_train.groupby('Pclass')
```
