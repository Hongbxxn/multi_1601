```python
def print_numbers(*args):
    for arg in args:
        print(arg)

print_numbers(10)
print('-----------')
print_numbers(10,20,30)
x=[1,2,3,4,5,6]
print_numbers(*x)
```

```python
def print_numbers(a, *args):
    print(a)
    print(args)

print_numbers(10)
print('-----------')
print_numbers(10,20,30)
x=[1,2,3,4,5,6]
print_numbers(*x)
```

```python
def person_info(name, age, address):
    print('이름 :' , name)
    print('나이 :',age)
    print('주소 :',address)

person_info('강경미',20,'경기도 고양시 일산동구')
```

```python
def person_info(**kwargs):
    for kw, arg in kwargs.items():
        print(kw,':',arg,sep=' ')
```

```python
def hello(count):
    if count == 0:
        return 
    print('hello world', count)
    count -= 1
    hello(count)
hello(5)
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(3))
```
